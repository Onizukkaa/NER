```python
#Open the text file
text_file=open("Natural_language_Processing_Text.txt")
#Read the data
text=text_file.read()
#Datatype of the data read
print(type(text))
print("\n")
#print the text
print(f"Visualisation du texte : \n {text}")
print("\n")
#length of the text
print(f"Longueur du texte : {len(text)}")
```

```python
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
nltk.download("popular")
```

```python
#tokenize the text by snetences:
sentences=sent_tokenize(text)
#how manu sentences are ther?
print(f"Nombre de sentences : {len(sentences)}")
print("\n")
#print sentences
print(f"Visualisation des sentences :\n {sentences}")
```

```python
#tokenize the text with words:
words=word_tokenize(text)
print(f"Nombre de mots : {len(words)}")
print("\n")
print(f"Visualisation des mots : \n {words}")
```

```python
from nltk.probability import FreqDist
#find the frequency
fdist=FreqDist(words)
#print  most common words

print(f"Les 10 mots les plus fréquents: \n {fdist.most_common(10)}")
```

```python
#plot the graph for fdist
import matplotlib.pyplot as plt
fdist.plot(10)
```

```python
#Empty list store words
words_no_punc=[]
#removing punctuation marks
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())
        
print(f"Mots sans ponctuations :\n {words_no_punc}")
print(f"Il y'a {len(words_no_punc)} mots sans ponctuation")
```

```python
#fréquency distrib
fdist=FreqDist(words_no_punc)
print(f"10 premiers mots sans ponctuation et leur distribution: \n {fdist.most_common(10)}")
```

```python
fdist.plot(10)
```

```python
from nltk.corpus import stopwords
#list of stopwords
stopwords=stopwords.words("english")
print(f"les stopwords : \n {stopwords}")
```

```python
#nettoyage 
clean_words=[]
for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)
        
print(f"mots clean :\n {clean_words}")
print(f"Total de clean words : {len(clean_words)}")
```

```python
#fréquency distrib
fdist=FreqDist(clean_words)
print(f"10 premiers mots clean et leur distribution: \n {fdist.most_common(10)}")
```

```python
fdist.plot(10)
```

```python
from wordcloud import WordCloud
#Generating the wordcloud
wordcloud=WordCloud().generate(text)

plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
#to remove the axis value
plt.axis("off")
plt.show()
```

```python
import numpy as np
from PIL import Image

#use a circle as mask
char_mask=np.array(Image.open("circle.png"))
#Generating the wordcloud
wordcloud=WordCloud(background_color="black",mask=char_mask).generate(text)

plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
#to remove the axis value
plt.axis("off")
plt.show()

```

```python
#stemming
from nltk.stem import PorterStemmer
porter=PorterStemmer()

word_list=clean_words

for w in word_list:
    print(porter.stem(w))
```
